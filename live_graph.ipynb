```python
%%time

losses=[]

pbar = tqdm(total=20)
for i in range(20):
    losses.append(np.sin(i))
    
    plt.plot(np.arange(len(losses)), losses)
    plt.show()    

#     display.clear_output(wait=True)
    
    pbar.update()
```
